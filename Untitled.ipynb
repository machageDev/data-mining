```python
import pandas as pd
df1 = pd.read_csv("/winequality-red.csv",sep=';')
df2 = pd.read_csv("/winequality-white.csv",sep=';')
df1.head()
df2.head()
df_all = pd.merge(df1,df2)
df_all.head()
df_all = pd.concat([df1,df2],axis=0)
df_all.head()
df_all.isnull().sum()
df_all.drop_duplicates(inplace=True)
df_all.info()
df_all.describe()
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df_all['quality'], kde=True)
plt.title("Distribution of Wine Quality")
plt.show()
sns.boxplot(x='quality', y='alcohol', data=df_all)
plt.title("Boxplot: Alcohol Content vs Wine Quality")
plt.show()
sns.violinplot(x='quality', y='alcohol', data=df_all)
plt.title("Violinplot: Alcohol Content vs Wine Quality")
plt.show()
sns.violinplot(x="quality", y="alcohol", data=df_all)
plt.title("Violinplot: Alcohol Content vs Wine Quality")
plt.show()
sns.histplot(x='quality', data=df_all)
plt.title("Distribution of Wine Types (Red vs White)")
plt.show()
```

```python
x = df1.drop('quality', axis=1)
y = df1['quality']
```

```python
y = df1['quality']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_predict)
print(r2)
```
